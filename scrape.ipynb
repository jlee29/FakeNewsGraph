```python
import urllib2
from bs4 import BeautifulSoup
import re
```

```python
page = urllib2.urlopen('https://mediabiasfactcheck.com/left/')
```

```python
soup = BeautifulSoup(page)
```

```python
relevantLinks = []
relevant = False
for link in soup.find_all('a'):
    if link.get('href') == 'https://mediabiasfactcheck.com/act-tv/':
        relevant = True
    
    if relevant:
        relevantLinks.append(link.get('href'))
        
    if link.get('href') == 'http://mediabiasfactcheck.com/z-magazine/':
        relevant = False
```

```python
urlLeftSet = set()
progress = 0
for link in relevantLinks:
    print(progress)
    progress += 1
    try:
        page = urllib2.urlopen(link)
    except:
        continue
    soup = BeautifulSoup(page).encode_contents(formatter='html')
    beginIndex = soup.find("Source:")
    endIndex = soup[beginIndex:].find('\n')
    line = soup[beginIndex:beginIndex+endIndex]
    urls = re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', line)
    if len(urls) > 0:
        urlLeftSet.add(urls[0])
print urlLeftSet
```
